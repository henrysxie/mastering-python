```python
df = pd.read_csv(
    '../data/chicago-salaries.csv',
    converters={
        'Employee Annual Salary': lambda x: float(x.replace('$', ''))
    })
```

```python
df
```

```python
df = df.rename(columns={'Name': 'name',
                       'Position Title': 'title',
                       'Department': 'dept',
                       'Employee Annual Salary': 'salary'})
```

```python
df
```

```python
df['salary'].describe()
```

```python
dept_sizes = df.groupby('dept').size()
```

```python
dept_sizes.plot(kind='barh', figsize=(15, 20))
```

```python
df['dept'].value_counts()
```

```python
df.groupby('dept').apply(lambda dept: len(dept['title'].unique()))
```

```python
df.groupby('dept').apply(
    lambda dept: dept['title'].apply(len).mean()
).plot(kind='barh', figsize=(15,15))
```

```python
df.groupby('dept').apply(
    lambda dept: max(dept['salary']) - min(dept['salary'])
).order(ascending=False)
```

```python
df[df['dept'] == 'POLICE'].sort('salary')
```

```python
df.groupby('dept').apply(
    lambda dept: dept['salary'].max()
).order(ascending=False)
```
