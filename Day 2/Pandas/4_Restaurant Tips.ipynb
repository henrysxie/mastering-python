```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = pd.read_csv('../data/tips.csv')
```

```python
tips.head()
```

```python
tips['tip_pct'] = tips['tip'] / tips['total_bill']
```

```python
tips
```

```python
grouped_pct = tips.groupby(['sex', 'smoker'])
```

```python
grouped_pct.agg(['mean', 'std'])
```

```python
def top(df, color, n=5, column="tip_pct"):
    return df.sort(column, ascending=False)[:n]
```

```python
top(tips, n=3)
```

```python
tips.groupby('smoker').apply(top)
```

```python
tips.groupby(['smoker', 'day'])['total_bill'].max()
```

```python
tips.groupby(['smoker', 'day']).apply(top, 'red', n=1, column="total_bill")
```

```python
table = tips.pivot_table(['size'], index=['day'], columns=['sex', 'smoker'], aggfunc='sum', margins=True)
```

```python
party_counts = pd.crosstab(tips.day, tips['size'])
```

```python
party_counts
```

```python
party_counts = party_counts.ix[:,2:5]
```

```python
party_counts.plot(kind='bar')
```

```python
party_counts
```

```python
party_pcts = party_counts.div(party_counts.sum(1), axis=0)
```

```python
party_pcts
```

```python
party_pcts.plot(kind='bar', stacked=True)
```
